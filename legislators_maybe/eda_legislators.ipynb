```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load the dataset
df = pd.read_csv('cleaned_legislators.csv')

# Display basic information about the dataset
print("Dataset Info:")
print(df.info())

# Display the first few rows
print("\nFirst 5 Rows:")
print(df.head())
```

```python
# Summary statistics
print("\nSummary Statistics:")
print(df.describe())
```

```python
# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Set style for better visualizations
plt.style.use('fivethirtyeight')
sns.set_palette("husl")

# 1. Party Distribution
plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 1)
party_counts = df['party'].value_counts()
plt.pie(party_counts.values, labels=party_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Political Parties')

# 2. Age Distribution at Start of Term
plt.subplot(2, 2, 2)
sns.histplot(data=df, x='age_start', bins=20, kde=True)
plt.axvline(df['age_start'].mean(), color='red', linestyle='--', label='Mean Age')
plt.title('Age Distribution at Start of Term')
plt.xlabel('Age')
plt.legend()

# 3. Term Duration by Position Type
plt.subplot(2, 2, 3)
sns.boxplot(data=df, x='type', y='duration_years')
plt.title('Term Duration by Position Type')
plt.xlabel('Position Type')
plt.ylabel('Years in Office')

# 4. Zodiac Sign Distribution
plt.subplot(2, 2, 4)
zodiac_counts = df['zodiac_sign'].value_counts()
plt.bar(zodiac_counts.index, zodiac_counts.values)
plt.xticks(rotation=45)
plt.title('Distribution of Zodiac Signs')

plt.tight_layout()
plt.show()
```
