```python
import pandas as pd

import ast
```

```python
legislators_current = pd.read_csv('/Users/kayliemendoza/Desktop/COGS108 Final Project/csv_output/legislators-current.csv')
legislators_historical = pd.read_csv('/Users/kayliemendoza/Desktop/COGS108 Final Project/csv_output/legislators-historical.csv')

legislators = pd.concat([legislators_current, legislators_historical], ignore_index=True)
```

```python
from legislators_maybe.extract_legislators_info import extract_legislators_info

# Apply the extract_legislators_info function
legislators_extracted = legislators.apply(lambda row: pd.Series(extract_legislators_info(row)), axis=1)

# Combine the extracted columns with the original DataFrame
legislators = pd.concat([legislators, legislators_extracted], axis=1)
```

```python
from functions.get_zodiac_sign import get_zodiac_sign

# Apply the zodiac sign function to the 'birthday' column
legislators['zodiac_sign'] = legislators['birthday'].apply(get_zodiac_sign)
```

```python
# Import the get_zodiac_color function
from functions.zodiac_color_mapping import get_zodiac_color

# Add a new column for zodiac colors in the executive DataFrame
legislators['zodiac_color'] = legislators['zodiac_sign'].apply(get_zodiac_color)
```

```python
from test_calculate_term_ages import calculate_term_ages_vectorized

# Apply the vectorized function to calculate ages
ages = calculate_term_ages_vectorized(legislators)

# Combine the calculated ages with the original DataFrame
legislators = pd.concat([legislators, ages], axis=1)
```

```python
from functions.calculate_term_duration import calculate_term_duration

# Apply the function to calculate term durations
term_durations = legislators.apply(lambda row: pd.Series(calculate_term_duration(row)), axis=1)

# Combine the calculated durations with the original DataFrame
legislators = pd.concat([legislators, term_durations], axis=1)
```

```python
from functions.create_full_name import create_full_name 

# Apply the function to create the 'full_name' column
legislators['full_name'] = legislators.apply(create_full_name, axis=1)
```

```python
columns_to_keep = ['full_name', 'birthdate', 'birthday', 'gender', 'type', 'start_term', 'end_term', 'state', 'class', 'district', 'party', 'how', 
'zodiac_sign', 'age_start', 'age_end', 'duration_days', 'duration_years', 'duration_months']

# Create a new DataFrame with only the specified columns
filtered_legislators = legislators[columns_to_keep]
```
